# EDA of SEC 10-K Financial Data

This notebook explores the annual financial data extracted from the SEC 10-K filings for **Microsoft**, **Tesla**, and **Apple**.

## Objectives

- Load and combine the three company CSV files.
- Inspect the structure, quality, and summary statistics of the data.
- Calculate year-over-year (YoY) percentage changes for each financial metric.
- Compare patterns across companies and fiscal years using groupings and aggregate functions.
- Summarize the key findings directly in the notebook using markdown narrative.

## Notes

- All monetary values are stored in **millions of USD**.
- Each company uses its own fiscal calendar, so fiscal years should be interpreted in the context of the corresponding filing period.

## 1. Load and Prepare the Data

We load each company-specific CSV, standardize the combined dataset, and sort the rows so that time-based analysis is reproducible and easy to read.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

base_path = Path(".")
files = [
    base_path / "microsoft_10k_financials_2023_2025.csv",
    base_path / "tesla_10k_financials_2023_2025.csv",
    base_path / "apple_10k_financials_2023_2025.csv",
]

frames = [pd.read_csv(file_path) for file_path in files]
df = pd.concat(frames, ignore_index=True)
df["period_end"] = pd.to_datetime(df["period_end"])
df = df.sort_values(["company", "fiscal_year"]).reset_index(drop=True)

metric_cols = [
    "total_revenue_musd",
    "net_income_musd",
    "total_assets_musd",
    "total_liabilities_musd",
    "cash_flow_from_operating_activities_musd",
]

df
```

## 2. Initial Data Inspection

Before computing trends, we inspect the shape, data types, missing values, and overall distributions. This is a standard first pass in EDA because it reveals whether the dataset is clean enough for comparison across companies and years.

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print("\nData types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isna().sum())

display(df.head())
display(df[metric_cols].describe().T)
```

## 3. Descriptive Statistics and Grouped Aggregation

Next, we use aggregate functions to understand the data at two levels:

- **By company**: compare average, minimum, maximum, and variability across the three filings.
- **By fiscal year**: compare the companies on the same year axis to see how the group evolves over time.

```python
company_summary = df.groupby("company")[metric_cols].agg(["mean", "min", "max", "std"]).round(2)
year_summary = df.groupby("fiscal_year")[metric_cols].agg(["mean", "min", "max"]).round(2)

display(company_summary)
display(year_summary)
```

## 4. Year-over-Year Trend Analysis

For each company, we calculate YoY percentage change for the five financial metrics. This highlights whether growth or decline is accelerating, moderating, or reversing from one fiscal year to the next.

The formula used is:

\[
\text{YoY \% Change} = \frac{\text{Current Year} - \text{Previous Year}}{\text{Previous Year}} \times 100
\]

A first-year observation for each company will naturally have `NaN` for YoY change because there is no prior year in the dataset.

```python
df = df.sort_values(["company", "fiscal_year"]).copy()

for column in metric_cols:
    df[f"{column}_yoy_pct"] = df.groupby("company")[column].pct_change() * 100

yoy_columns = [f"{column}_yoy_pct" for column in metric_cols]
yoy_view = df[["company", "fiscal_year"] + metric_cols + yoy_columns].copy()

display(yoy_view.round(2))

yoy_summary = df.groupby("company")[yoy_columns].mean().round(2)
display(yoy_summary)
```

## 5. Cross-Company and Cross-Year Comparisons

To explore the data across different dimensions, we reshape it into simple pivot tables and examine correlations among the financial metrics. This helps us compare companies within the same fiscal year and see which metrics tend to move together.

```python
revenue_pivot = df.pivot(index="fiscal_year", columns="company", values="total_revenue_musd").round(2)
assets_pivot = df.pivot(index="fiscal_year", columns="company", values="total_assets_musd").round(2)
liabilities_pivot = df.pivot(index="fiscal_year", columns="company", values="total_liabilities_musd").round(2)

display(revenue_pivot)
display(assets_pivot)
display(liabilities_pivot)
display(df[metric_cols].corr().round(3))
```

## 6. Visualization

The line charts below provide a compact visual view of how each company evolves over time for the five financial metrics. Because the sample is small, line charts are a better fit than more complex statistical graphics.

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
axes = axes.flatten()

plot_metrics = metric_cols + [None]
for axis, column in zip(axes, plot_metrics):
    if column is None:
        axis.axis("off")
        continue

    for company_name, company_frame in df.groupby("company"):
        axis.plot(
            company_frame["fiscal_year"],
            company_frame[column],
            marker="o",
            linewidth=2.5,
            label=company_name,
        )

    axis.set_title(column.replace("_", " ").title())
    axis.set_xlabel("Fiscal Year")
    axis.set_ylabel("USD (millions)")
    axis.legend()

plt.tight_layout()
plt.show()
```

## 7. Findings and Conclusion

### Methodology Recap

- Loaded the three company CSV files and combined them into one analysis table.
- Checked the dataset structure, types, and missing values.
- Used grouped aggregations to compare companies and fiscal years.
- Calculated YoY percentage changes for all five financial metrics.
- Visualized the trends to make year-to-year movement easier to interpret.

### Observations

- **Microsoft** shows steady growth across revenue, net income, assets, and operating cash flow over the three-year window.
- **Tesla** has relatively smaller profits and more volatility in net income compared with its revenue base, while assets and liabilities continue to expand.
- **Apple** remains the largest revenue and profitability profile in absolute terms among the three companies in this dataset, with comparatively stable operating cash flow.
- Across all three companies, total assets and total liabilities generally move upward over time, which is consistent with business scale expansion and balance-sheet growth.

### Conclusion

This EDA shows that the companies follow different financial patterns even though they are all large-cap technology firms. Microsoft and Apple appear more stable and consistently profitable across the period, while Tesla exhibits stronger year-to-year variability in earnings. The YoY analysis and grouped summaries make those differences visible in a compact, reproducible way.
